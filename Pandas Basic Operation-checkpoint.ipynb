# Performing Basic Pandas operation

# Loading Data into Pandas

```python
import pandas as pd
```

```python
data = pd.read_csv('irisdata.csv')
```

### Head of data

```python
data.head(3)
```

Here column name are not specify so we need to give name to columns.

### Naming columns

```python
data.columns = ['SepalLengthCm','SepalWidthCm','PetalLenghtCm','PetalWidth','Species']
```

```python
data.head()
```

### tail of data

```python
data.tail()
```

### Shape of data 

```python
data.shape
```

### Iteration through rows

```python
data.iloc[:5,:4]
```

 :5 this means 1st 5 rows, :4 this means 1st 4 columns

```python
data.index
```

This gives information about index of dataset

```python
data.iloc[3:6,:-1]
```

```python
data.iloc[5:8,1:3]
```

### Getting rows by specific condition

```python
data.loc[data['SepalLengthCm'] >= 5.0]
```

```python
data.loc[data['Species'] == 'Iris-virginica']
```

### Basic discription of data

```python
data.describe()
```

This shows count, mean, standard deviation, minimum and maximum value, 25%, 50%, 75% of data

```python
data.describe(include=['object'])
```

By passing include=['object'] shows information about catogorical variable

### Sorting Data

```python
data.sort_values('SepalLengthCm', ascending=False)
```

```python
data.sort_values(['SepalLengthCm', 'PetalLenghtCm'], ascending=False)
```

### Aggregate statistics

```python
data.groupby(['Species']).mean()
```

Mean value of each columns groupby Species

```python
data.groupby(['Species']).sum()
```

Sum of each columns groupby Species

```python
data.groupby(['Species']).count()
```
